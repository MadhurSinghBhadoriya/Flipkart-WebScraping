##  Data Scraping of Footballs from Flipkart 

```python
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager

driver = webdriver.Firefox(executable_path=GeckoDriverManager().install())

import pandas as pd
import time
import os
```

```python
# Flipkart link

driver.get('https://www.flipkart.com/sports/football/pr?sid=abc,gxg&p[]=facets.serviceability%5B%5D%3Dtrue&otracker=categorytree&otracker=nmenu_sub_Sports%2C%20Books%20%26%20More_0_Football&otracker=nmenu_sub_Sports%2C%20Books%20%26%20More_0_Football')
```

```python
# FLIPKART
# """     
names = driver.find_elements_by_class_name('s1Q9rs')
offerpr = driver.find_elements_by_class_name('_30jeq3')
offerDetails = driver.find_elements_by_class_name('_3Ay6Sb')
originalPrice = driver.find_elements_by_class_name('_3I9_wc')


nameList = []
offerpriceList = []
discList = []
actualpriceList = []

for i in range(len(names)):
    
    name = names[i].text
    offerprice = offerpr[i].text
    disc = offerDetails[i].text
    actualprice = originalPrice[i].text
    
    
    nameList.append(name)
    offerpriceList.append(offerprice)
    discList.append(disc)
    actualpriceList.append(actualprice)
    
    
# intialise data of lists.
data = {'Brand Name/Size':nameList,
        'Offer Price': offerpriceList,
        'Discount': discList,
        'Actual Price': actualpriceList,
        
       }

# Create DataFrame
df = pd.DataFrame(data)
# """
```
